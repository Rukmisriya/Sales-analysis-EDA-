-------SALES ANALYSIS--------

```python
# Productline wise sales
product_sales = df.groupby('Product line')['Total'].sum().sort_values(ascending=False)
print(product_sales)
```

```python
# Identify best-selling product lines
print("Best-Selling Product Lines:")
print(product_sales.head())

# Identify least-selling product lines
print("\nLeast-selling Product Lines:")
print(product_sales.tail())
```

```python
#Product line sales analysis
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Product line', y='Total', palette='Set2')
plt.title('Sales by Product Line')
plt.show()
```

```python
#Dividing the customers
# sales analysis by Customer type
customer_type_sales = df.groupby('Customer type')['Total'].sum()
print("Sales by Customer Type:")
print(customer_type_sales)

# sales analysis by Gender
gender_sales = df.groupby('Gender')['Total'].sum()
print("\nSales by Gender:")
print(gender_sales)
```

```python
# Monthly sales analysis
month_sales = df.groupby('Month')['Total'].sum()
plt.figure(figsize=(12, 6))
month_sales.plot(kind='line', color=['orange'])
plt.title('Monthly sales Analysis')
plt.xlabel('Months')
plt.ylabel('Sales')
```

```python
# City sales Analysis relative to each and other
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='City', y='Total', palette='Set3')
plt.title('Sales Distribution by City')
plt.show()
```

```python
# Every day sales analysis
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Day', y='Total', palette='Set3')
plt.title('Sales Distribution by City')
plt.show()
```

```python
# relation between product line and gender
df.groupby(['Product line', 'Gender']).size().unstack()
```

```python
plt.figure(figsize=(15, 10))
df1=df.groupby(['Product line', 'Gender']).size().unstack()
df1.plot(kind='bar')
plt.title('Product Line - Gender Analysis')
plt.legend()
plt.show()
```

```python
# Product line performance by branch
df.groupby(['Product line', 'Branch'])['Total'].sum().unstack()
```

```python
# best-performing branch for each product line
plt.figure(figsize=(15, 10))
df1=df.groupby(['Product line', 'Branch'])['Total'].sum().unstack()
df1.plot(kind='bar')
plt.title('Product Line - Branch Analysis')
plt.legend()
plt.show()
```

```python
# Finding products with lower ratings
low_rated = df[df['Rating'] < 6] 
print("\nProducts with Ratings Below 6:")
print(low_rated[['Product line', 'Total', 'Rating']].head())
```

```python
# daily sales profit
daily_profit=df.groupby(df['Date'])['gross income'].sum().reset_index()
```

```python
plt.figure(figsize=(10,6))
plt.plot(daily_profit[['Date']], daily_profit['gross income'], marker='o', linestyle='-', color='skyblue')
plt.title('Daily Sales Profit Analysis')
plt.xlabel('Date')
plt.ylabel('Gross Income (Profit)')
plt.show()
```
